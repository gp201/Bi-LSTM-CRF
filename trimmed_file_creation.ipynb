```python
import pandas as pd
data = []
tags = []
text = True
fr = open("data/labelled_data.txt")
for i in fr.readlines():
    if text:
        txt = i.split()
        text = not text
    else:
        typ = i.split()
        tmp = [[x,y] for (x,y) in zip(txt,typ)]
        data.extend(tmp)
        text = not text
df = pd.DataFrame(data,columns=["word","tags"])
```

```python
embeddings_vocab = pd.read_csv("data/vocab.tsv", sep="\t", names=["word", "occurences"]).reset_index()
embeddings_vocab
```

```python
merge_df = df.merge(embeddings_vocab, on="word")
merge_df.drop_duplicates(["index","word", "occurences"], inplace=True)
merge_df
```

```python
# Write trimmed vocab
f = open("data/vocab.tsv")
fout = open("data/trimmed_vocab.tsv", "w")

indices = merge_df["index"].tolist()
for ctr, i in enumerate(f.readlines()):
    if ctr %10 == 0:
        print(ctr)
    if ctr in indices:
        fout.write(i)

fout.close()
f.close()

```

```python
# Write trimmed vectors
f = open("data/vectors.tsv")
fout = open("data/trimmed_vectors.tsv", "w")

indices = merge_df["index"].tolist()
for ctr, i in enumerate(f.readlines()):
    if ctr %10 == 0:
        print(ctr)
    if ctr in indices:
        fout.write(i)

fout.close()
f.close()
```
